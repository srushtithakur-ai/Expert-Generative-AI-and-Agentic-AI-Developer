# All / Any
- the all() method returns : true if all the elements in the list is true ( not 0 )
- the all() method returns : false if any of the element in the list is false (includes 0)
- the any() function returns true if any element in the list is true if not any() returns false 

```python
l1 = [1,2,3,4,0]
l1
```

```python
all(l1) #the list has 0 in it so it shows false 
```

```python
any(l1) #will print true because we have other items which are true in value 
```

```python
l2 = [1,2,3,4,5,True,False]  
```

```python
all(l2) #the list has false in it so it shows false
```

```python
any(l2) #will print true because we have other items which are true in value 
```

```python
l3 = [1,2,3,4,5,True]  
```

```python
all(l3) #return as true because all the values in the list are true 
```

```python
any(l3) #return as true because all the values in the list are true 
```

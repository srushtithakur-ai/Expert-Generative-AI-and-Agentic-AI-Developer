# 3. Python Data Structures
- there are 2 types of data structure
- a. inbuild or build-in , b. user define
- inbuilt : list, tuple, set, dict, range
- user defined: array, linked list, tree
- user defined data structure will be taken care by library or package or frameworks.
- in data type user define only 1 value eg: a=5, b=5.5, c='Srushti'
      . int, float, bool, complex and string
- in data structure the user define more than 1 value eg b=[2,3,4,5], c=('a','b',3.5)

- Matrix is also called array
- A general container for multiple values is called array, Can be 1D, 2D, 3D, ... ND.
      . 1d, 2d, nd, transpose, flatten etc
- Dataframe is called table(row*column)

- Tensor collections of array or matrix
- array is collection of vector
